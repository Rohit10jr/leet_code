**88. Merge Sorted Array**

You are given two integer arrays nums1 and nums2, sorted in non-decreasing order, and two integers m and n, representing the number of elements in nums1 and nums2 respectively.

Merge nums1 and nums2 into a single array sorted in non-decreasing order.

The final sorted array should not be returned by the function, but instead be stored inside the array nums1. To accommodate this, nums1 has a length of m + n, where the first m elements denote the elements that should be merged, and the last n elements are set to 0 and should be ignored. nums2 has a length of n.

 

Example 1:

Input: nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3
Output: [1,2,2,3,5,6]
Explanation: The arrays we are merging are [1,2,3] and [2,5,6].
The result of the merge is [1,2,2,3,5,6] with the underlined elements coming from nums1.

Approach 1: Two Pointers (From the End) â€” Optimal

-  Time Complexity: O(m + n)
-  Space Complexity: O(1) (in-place)

```python
class Solution:
    def merge_three_pointers_from_end(self, nums1, m, nums2, n):
        # Merges arrays in-place from the end to avoid overwriting, leveraging extra space in nums1. Time: O(m + n), Space: O(1)
        i, j, k = m - 1, n - 1, m + n - 1
        i, j, k = m - 1, n - 1, m + n - 1

        print(f"Initial nums1, m, nums2, n: {nums1}, {m}, {nums2}, {n}")
        print(f"Initial values of i, j, k: {i}, {j}, {k}")
        print("Starting merge...\n")

        while i >= 0 and j >= 0:
            print(f"Comparing nums1[{i}]={nums1[i]} and nums2[{j}]={nums2[j]}")
            if nums1[i] > nums2[j]:
                nums1[k] = nums1[i]
                print(f"Placing nums1[{i}] at nums1[{k}]")
                i -= 1
            else:
                nums1[k] = nums2[j]
                print(f"Placing nums2[{j}] at nums1[{k}]")
                j -= 1
            k -= 1
            print(f"nums1 now: {nums1}\n")

        # If any elements are left in nums2
        while j >= 0:
            print(f"Copying remaining nums2[{j}]={nums2[j]} to nums1[{k}]")
            nums1[k] = nums2[j]
            j -= 1
            k -= 1
            print(f"nums1 now: {nums1}\n")

        print("Merge complete.")
        print(f"Final nums1: {nums1}")


    def merge_copy_and_sort(self, nums1, m, nums2, n):
        # Copies nums2 into the empty part of nums1 and then sorts the entire array. Time: O((m+n) log(m+n)), Space: O(1) (in-place sort)
        nums1[m:] = nums2
        nums1.sort()
        print("Merged nums1:", nums1)

    def merge_overwrite_and_sort(self, nums1, m, nums2, n):
        for j in range(n):
            nums1[m + j] = nums2[j]
        nums1.sort()
        print("Merged nums1:", nums1)

    def merge_filter_and_sort(self, nums1, m, nums2, n):
        # Filters non-zero elements, then extends with nums2 and sorts the combined list. Time: O(m + n + (m+n) log(m+n)), Space: O(m)
        num = []
        # for i in range(len(nums1)):
            # if nums1[i] != 0:
                # num.append(nums1[i])
        # for i in nums:
        #     if i != 0:
        #         num.append(i)
        for i in nums1[:m]:
            num.append(i)
        nums1[:] = num
        nums1.extend(nums2)
        nums1.sort()
        print("Merged nums1:", nums1)

    def merge_in_place_with_insert(self, nums1, m, nums2, n):
        # Iterates through nums1 and inserts elements from nums2 in sorted order. Time: O(m*n), Space: O(1)
        nums1[:] = nums1[:m] # Remove trailing zeros 
        i, j = 0, 0
        while i < m and j < n:
            if nums1[i] <= nums2[j]:
                i += 1
            else:
                nums1.insert(i, nums2[j])
                j += 1
                m += 1
        if j < n:
            nums1.extend(nums2[j:])
        print("Merged nums1:", nums1)
```

```python
s = Solution()
nums1_base = [1, 2, 3, 0, 0, 0]
nums2_base = [2, 4, 5]
m_base = 3
n_base = 3

print(f"--- Calling all methods with example: nums1={nums1_base[:m_base]}, nums2={nums2_base} ---")

# Test merge_three_pointers_from_end (Optimal)
nums1_test1 = list(nums1_base) # Create a fresh copy
print(f"Method: merge_three_pointers_from_end | Initial: {nums1_test1}")
s.merge_three_pointers_from_end(nums1_test1, m_base, nums2_base, n_base)
print(f"Result : {nums1_test1}\n")

# Test merge_copy_and_sort (Good balance of simplicity and performance)
nums1_test2 = list(nums1_base) # Create a fresh copy
print(f"Method: merge_copy_and_sort | Initial: {nums1_test2}")
s.merge_copy_and_sort(nums1_test2, m_base, nums2_base, n_base)
print(f"Result : {nums1_test2}\n")

# Test merge_overwrite_and_sort (Similar to merge_copy_and_sort)
nums1_test_overwrite_sort = list(nums1_base) # Create a fresh copy
print(f"Method: merge_overwrite_and_sort | Initial: {nums1_test_overwrite_sort}")
s.merge_overwrite_and_sort(nums1_test_overwrite_sort, m_base, nums2_base, n_base)
print(f"Result : {nums1_test_overwrite_sort}\n")

# Test merge_filter_and_sort (Has a bug, and less efficient)
nums1_test3 = list(nums1_base) # Create a fresh copy
print(f"Method: merge_filter_and_sort | Initial: {nums1_test3}")
s.merge_filter_and_sort(nums1_test3, m_base, nums2_base, n_base)
print(f"Result : {nums1_test3}\n") # Will likely be incorrect due to `num.append(i)` bug

# Test merge_in_place_with_insert (Inefficient for large N)
nums1_test4 = list(nums1_base) # Create a fresh copy
print(f"Method: merge_in_place_with_insert | Initial: {nums1_test4}")
s.merge_in_place_with_insert(nums1_test4, m_base, nums2_base, n_base)
print(f"Result : {nums1_test4}\n")
```
