**27. Remove Element**

Given an integer array nums and an integer val, remove all occurrences of val in nums in-place. The order of the elements may be changed. Then return the number of elements in nums which are not equal to val.

Consider the number of elements in nums which are not equal to val be k, to get accepted, you need to do the following things:

Change the array nums such that the first k elements of nums contain the elements which are not equal to val. The remaining elements of nums are not important as well as the size of nums.
Return k.

Example 1:

Input: nums = [3,2,2,3], val = 3
Output: 2, nums = [2,2,_,_]

Explanation: Your function should return k = 2, with the first two elements of nums being 2.
It does not matter what you leave beyond the returned k (hence they are underscores).
Example 2:

Input: nums = [0,1,2,2,3,0,4,2], val = 2
Output: 5, nums = [0,1,4,0,3,_,_,_]

Explanation: Your function should return k = 5, with the first five elements of nums containing 0, 0, 1, 3, and 4.
Note that the five elements can be returned in any order.
It does not matter what you leave beyond the returned k (hence they are underscores).

```python
class Solution(object):

    def removeElement_two_pointers(self, nums, val):
        # Moves non-matching elements to the front using two pointers. Time: O(N), Space: O(1)
        k = 0
        for i in range(len(nums)):
            if nums[i] != val:
                nums[k] = nums[i]
                k += 1
        return k

    def removeElement_two_pointer_swap(self, nums, val):
        # Swaps target elements from the front with non-target elements from the back. Time: O(N), Space: O(1)
        i = 0
        j = len(nums) - 1
        while i <= j:
            if nums[i] == val:
                nums[i] = nums[j]
                j -= 1
            else:
                i += 1
        return i

    def removeElement_list_comprehension(self, nums, val):
        # Creates a new list excluding target elements and assigns it back in-place. Time: O(N), Space: O(N)
        nums[:] = [num for num in nums if num != val]
        return len(nums)

    def removeElement_reverse_loop(self, nums, val):
        # Iterates backward and removes target elements using list.pop(). Time: O(N^2), Space: O(1)
        for i in range(len(nums) - 1, -1, -1):
            if nums[i] == val:
                nums.pop(i)
        return len(nums)

    def removeElement_while_loop(self, nums, val):
        # Iterates forward and removes target elements using list.pop(). Time: O(N^2), Space: O(1)
        i = 0
        while i < len(nums):
            if nums[i] == val:
                nums.pop(i)
            else:
                i += 1
        return len(nums)
```

```python

# Example Usage:
sol = Solution()
val_to_remove = 3

# Test removeElement_two_pointers
nums_tp = [3, 2, 2, 3]
original_nums_tp = list(nums_tp) # Create a copy for comparison
length_tp = sol.removeElement_two_pointers(nums_tp, val_to_remove)
print(f"removeElement_two_pointers:")
print(f"  Original: {original_nums_tp}")
print(f"  Modified: {nums_tp[:length_tp]} (length: {length_tp})")
print(f"  Full array after modification (elements beyond new length don't matter): {nums_tp}\n")


# Test removeElement_two_pointer_swap
nums_tps = [0, 1, 2, 2, 3, 0, 4, 2]
original_nums_tps = list(nums_tps) # Create a copy
length_tps = sol.removeElement_two_pointer_swap(nums_tps, val_to_remove)
print(f"removeElement_two_pointer_swap:")
print(f"  Original: {original_nums_tps}")
print(f"  Modified: {nums_tps[:length_tps]} (length: {length_tps})")
print(f"  Full array after modification (elements beyond new length don't matter): {nums_tps}\n")


# Test removeElement_list_comprehension
nums_lc = [3, 2, 2, 3]
original_nums_lc = list(nums_lc) # Create a copy
length_lc = sol.removeElement_list_comprehension(nums_lc, val_to_remove)
print(f"removeElement_list_comprehension:")
print(f"  Original: {original_nums_lc}")
print(f"  Modified: {nums_lc[:length_lc]} (length: {length_lc})")
print(f"  Full array after modification: {nums_lc}\n")


# Test removeElement_reverse_loop
nums_rl = [3, 2, 2, 3]
original_nums_rl = list(nums_rl) # Create a copy
length_rl = sol.removeElement_reverse_loop(nums_rl, val_to_remove)
print(f"removeElement_reverse_loop:")
print(f"  Original: {original_nums_rl}")
print(f"  Modified: {nums_rl[:length_rl]} (length: {length_rl})")
print(f"  Full array after modification: {nums_rl}\n")


# Test removeElement_while_loop
nums_wl = [0, 1, 2, 2, 3, 0, 4, 2]
original_nums_wl = list(nums_wl) # Create a copy
length_wl = sol.removeElement_while_loop(nums_wl, val_to_remove)
print(f"removeElement_while_loop:")
print(f"  Original: {original_nums_wl}")
print(f"  Modified: {nums_wl[:length_wl]} (length: {length_wl})")
print(f"  Full array after modification: {nums_wl}\n")
```
