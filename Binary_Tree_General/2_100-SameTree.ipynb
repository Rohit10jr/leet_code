**100. Same Tree**

Given the roots of two binary trees p and q, write a function to check if they are the same or not.
Two binary trees are considered the same if they are structurally identical, and the nodes have the same value.

Example 1:

    Input: p = [1,2,3], q = [1,2,3]
    Output: true

Example 2:

    Input: p = [1,2], q = [1,null,2]
    Output: false

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def isSameTree(self, p: TreeNode, q: TreeNode) -> bool:
        print("-"*30)
        print("p.val" if p else "p is None", p.val if p else "")
        print("q.val" if q else "q is None", q.val if q else "")
        print("-"*30)

        # Case 1: both are None (empty) → same
        if not p and not q:
            return True
        
        # Case 2: only one is None, or values differ → not same
        if not p or not q or p.val != q.val:
            return False
        
        # Recursive Step: Check if the left subtrees are the same
        # AND if the right subtrees are the same.
        return (
            self.isSameTree(p.left, q.left) and
            self.isSameTree(p.right, q.right)
        )
        # The backslash \ in Python lets you split one logical line of code across multiple physical lines to improve readability
        # return self.isSameTree(p.left, q.left) and \
            #    self.isSameTree(p.right, q.right)
```

```python
# Tree p
p = TreeNode(1)
p.left = TreeNode(2)
p.right = TreeNode(3)

# Tree q
q = TreeNode(1)
q.left = TreeNode(2)
q.right = TreeNode(3)

sol = Solution()
print(sol.isSameTree(p, q)) # Output: True
```
