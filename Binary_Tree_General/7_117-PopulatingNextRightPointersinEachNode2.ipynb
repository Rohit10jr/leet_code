**117. Populating Next Right Pointers in Each Node II**

Given a binary tree

struct Node {
  int val;
  Node *left;
  Node *right;
  Node *next;
}
Populate each next pointer to point to its next right node. If there is no next right node, the next pointer should be set to NULL.

Initially, all next pointers are set to NULL

Example 1:

    Input: root = [1,2,3,4,5,null,7]
    Output: [1,#,2,3,#,4,5,7,#]

Explanation: Given the above binary tree (Figure A), your function should populate each next pointer to point to its next right node, just like in Figure B. The serialized output is in level order as connected by the next pointers, with '#' signifying the end of each level.

Example 2:

    Input: root = []
    Output: []

```python
# Definition for a Node.
class Node:
    def __init__(self, val: int = 0, left: 'Node' = None, right: 'Node' = None, next: 'Node' = None):
        self.val = val
        self.left = left
        self.right = right
        self.next = next

class Solution:
    def connect(self, root: 'Node') -> 'Node':
        if not root:
            return None

        # current_level_start is the leftmost node of the level we are currently processing.
        current_level_start = root

        # We iterate level by level. This outer loop processes one level at a time.
        while current_level_start:
            # dummy_head helps us build the 'next' pointers for the *next* level.
            # Its 'next' will point to the very first node of the next level.
            dummy_head = Node(0) 
            
            # tail always points to the last node that was linked in the *next* level.
            # We append children of current_level_start's nodes to tail.next.
            tail = dummy_head

            # p is a pointer that traverses horizontally through the current level,
            # using the 'next' pointers that were set in the previous iteration.
            p = current_level_start 
            while p:
                # If current node 'p' has a left child, connect it to the 'next' level.
                if p.left:
                    tail.next = p.left
                    tail = tail.next # Advance tail
                
                # If current node 'p' has a right child, connect it to the 'next' level.
                if p.right:
                    tail.next = p.right
                    tail = tail.next # Advance tail
                
                # Move to the next node in the current level (using its 'next' pointer).
                p = p.next
            
            # After processing all nodes in the current level, move to the next level.
            # The head of the next level is dummy_head.next.
            current_level_start = dummy_head.next 
        
        return root
```

```python
# Build tree:
#        1
#      /   \
#     2     3
#    / \     \
#   4   5     7

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.right = Node(7)

sol = Solution()
print(sol.connect(root))
```
