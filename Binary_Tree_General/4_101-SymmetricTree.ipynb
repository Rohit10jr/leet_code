**101. Symmetric Tree**

Given the root of a binary tree, check whether it is a mirror of itself (i.e., symmetric around its center).

Example 1:

    Input: root = [1,2,2,3,4,4,3]
    Output: true
    
Example 2:

    Input: root = [1,2,2,null,3,null,3]
    Output: false

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def isSymmetric(self, root: TreeNode) -> bool:
        # Empty tree is symmetric
        if not root:
            return True
        
        # Compare left and right subtrees
        return self._isMirror(root.left, root.right)
    
    def _isMirror(self, node1: TreeNode, node2: TreeNode) -> bool:
        # Both nodes are None → symmetric
        if not node1 and not node2:
            return True
        
        # One is None or values differ → not symmetric
        if not node1 or not node2 or node1.val != node2.val:
            return False
            
        # Compare opposite children
        return (
            self._isMirror(node1.left, node2.right) and
            self._isMirror(node1.right, node2.left)
        )
```

```python
root = TreeNode(1)
root.left = TreeNode(2, TreeNode(3), TreeNode(4))
root.right = TreeNode(2, TreeNode(4), TreeNode(3))

sol = Solution()
print("Is Symmetric:", sol.isSymmetric(root))
```
