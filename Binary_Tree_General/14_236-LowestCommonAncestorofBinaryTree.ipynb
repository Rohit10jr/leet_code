**236. Lowest Common Ancestor of a Binary Tree**

Given a binary tree, find the lowest common ancestor (LCA) of two given nodes in the tree.

According to the definition of LCA on Wikipedia: “The lowest common ancestor is defined between two nodes p and q as the lowest node in T that has both p and q as descendants (where we allow a node to be a descendant of itself).”
 
Example 1:

    Input: root = [3,5,1,6,2,0,8,null,null,7,4], p = 5, q = 1
    Output: 3

Explanation: The LCA of nodes 5 and 1 is 3.

Example 2:

    Input: root = [3,5,1,6,2,0,8,null,null,7,4], p = 5, q = 4
    Output: 5

Explanation: The LCA of nodes 5 and 4 is 5, since a node can be a descendant of itself according to the LCA definition.

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        # Base case 1: If root is None, we didn't find p or q in this path.
        if not root:
            return None
        
        # Base case 2: If the current root is either p or q, then this root is
        # the LCA (since a node can be a descendant of itself).
        if root == p or root == q:
            return root
        
        # Recursively search in the left subtree.
        left_lca = self.lowestCommonAncestor(root.left, p, q)
        
        # Recursively search in the right subtree.
        right_lca = self.lowestCommonAncestor(root.right, p, q)
        
        # Now, combine the results from the left and right subtrees:

        # Case 1: Both p and q were found in different subtrees (one in left, one in right).
        # This means the current root is their LCA.
        if left_lca and right_lca:
            return root
        
        # Case 2: Only the left_lca is not None.
        # This means both p and q (or just one, but if so, that one is left_lca)
        # are in the left subtree. So, the LCA must be in the left subtree.
        elif left_lca:
            return left_lca
        
        # Case 3: Only the right_lca is not None.
        # This means both p and q (or just one, that one is right_lca)
        # are in the right subtree. So, the LCA must be in the right subtree.
        elif right_lca:
            return right_lca
        
        # Case 4: Neither left_lca nor right_lca is found.
        # This means p and q are not descendants of the current root.
        else:
            return None
```

```python
# --- Example Usage ---

def build_example_tree():
    # Tree structure for [3,5,1,6,2,0,8,null,null,7,4]
    #         3
    #        / \
    #       5   1
    #      / \ / \
    #     6  2 0  8
    #       / \
    #      7   4
    
    node3 = TreeNode(3)
    node5 = TreeNode(5)
    node1 = TreeNode(1)
    node6 = TreeNode(6)
    node2 = TreeNode(2)
    node0 = TreeNode(0)
    node8 = TreeNode(8)
    node7 = TreeNode(7)
    node4 = TreeNode(4)

    node3.left = node5
    node3.right = node1

    node5.left = node6
    node5.right = node2

    node1.left = node0
    node1.right = node8

    node2.left = node7
    node2.right = node4

    return node3, node5, node1, node4 # Return specific nodes needed for p and q

# 1. Initialize the tree and get references to p and q
root_node, node_p5, node_q1, node_q4 = build_example_tree()

# 2. Create an instance of the Solution class
solver = Solution()

# --- Test Case 1: p = 5, q = 1 ---
p_node1 = node_p5
q_node1 = node_q1

lca1 = solver.lowestCommonAncestor(root_node, p_node1, q_node1)
print(f"LCA of {p_node1.val} and {q_node1.val} is: {lca1.val}") # Expected: 3

print("-" * 30)

# --- Test Case 2: p = 5, q = 4 ---
p_node2 = node_p5
q_node2 = node_q4

lca2 = solver.lowestCommonAncestor(root_node, p_node2, q_node2)
print(f"LCA of {p_node2.val} and {q_node2.val} is: {lca2.val}") # Expected: 5

print("-" * 30)

# --- Test Case 3: p = 6, q = 7 (LCA should be 5) ---
# Need to get references to nodes 6 and 7 if not already done by build_example_tree
# Let's adjust build_example_tree or define them inline if needed
root_node_test3, node_p5_test3, node_q1_test3, node_q4_test3 = build_example_tree() # Rebuild for clarity

# Manually get node references for node6 and node7 for this test
# In a real scenario, you might have a map or a different tree builder.
node6 = root_node_test3.left.left
node7 = root_node_test3.left.right.left

lca3 = solver.lowestCommonAncestor(root_node_test3, node6, node7)
print(f"LCA of {node6.val} and {node7.val} is: {lca3.val}") # Expected: 2 (node2)
```
