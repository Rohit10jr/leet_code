**373. Find K Pairs with Smallest Sums**

You are given two integer arrays nums1 and nums2 sorted in non-decreasing order and an integer k.

Define a pair (u, v) which consists of one element from the first array and one element from the second array.

Return the k pairs (u1, v1), (u2, v2), ..., (uk, vk) with the smallest sums.


    Example 1:

    Input: nums1 = [1,7,11], nums2 = [2,4,6], k = 3
    Output: [[1,2],[1,4],[1,6]]
    Explanation: The first 3 pairs are returned from the sequence: [1,2],[1,4],[1,6],[7,2],[7,4],[11,2],[7,6],[11,4],[11,6]

    Example 2:

    Input: nums1 = [1,1,2], nums2 = [1,2,3], k = 2
    Output: [[1,1],[1,1]]
    Explanation: The first 2 pairs are returned from the sequence: [1,1],[1,1],[1,2],[2,1],[1,2],[2,2],[1,3],[1,3],[2,3]

```python
import heapq

class Solution:
    def kSmallestPairs(self, nums1, nums2, k):
        if not nums1 or not nums2:
            return []

        result = []
        min_heap = []

        # Step 1: push (nums1[0] + nums2[0], i=0, j=0)
        heapq.heappush(min_heap, (nums1[0] + nums2[0], 0, 0))
        print(f"Initial push → sum={nums1[0]+nums2[0]}, pair=({nums1[0]},{nums2[0]})")

        visited = set()
        visited.add((0, 0))

        iteration = 1

        while min_heap and len(result) < k:

            print("\n----------------------------")
            print(f"Iteration {iteration}")
            iteration += 1

            # Step 2: Pop smallest pair
            curr_sum, i, j = heapq.heappop(min_heap)
            print(f"POP → sum={curr_sum}, pair=({nums1[i]}, {nums2[j]})")

            result.append([nums1[i], nums2[j]])

            # Step 3: Push (i, j+1)
            if j + 1 < len(nums2) and (i, j+1) not in visited:
                heapq.heappush(
                    min_heap,
                    (nums1[i] + nums2[j+1], i, j+1)
                )
                visited.add((i, j+1))
                print(f"PUSH right → sum={nums1[i] + nums2[j+1]}, pair=({nums1[i]}, {nums2[j+1]})")

            # Step 4: Push (i+1, j) ONLY if j == 0
            if j == 0 and i + 1 < len(nums1) and (i+1, j) not in visited:
                heapq.heappush(
                    min_heap,
                    (nums1[i+1] + nums2[j], i+1, j)
                )
                visited.add((i+1, j))
                print(f"PUSH down → sum={nums1[i+1] + nums2[j]}, pair=({nums1[i+1]}, {nums2[j]})")

            # Show heap content
            print("Heap now contains:")
            print([ (s, i2, j2, f"({nums1[i2]},{nums2[j2]})") for (s,i2,j2) in min_heap ])

        return result
```

```python
# Create object
sol = Solution()

# Example input
nums1 = [1, 7, 11]
nums2 = [2, 4, 6]
k = 3

# Call the function
result = sol.kSmallestPairs(nums1, nums2, k)

print("K smallest pairs:", result)
```
