**103. Binary Tree Zigzag Level Order Traversal**

Given the root of a binary tree, return the zigzag level order traversal of its nodes' values. (i.e., from left to right, then right to left for the next level and alternate between).

Example 1:

    Input: root = [3,9,20,null,null,15,7]
    Output: [[3],[20,9],[15,7]]

Example 2:

    Input: root = [1]
    Output: [[1]]

```python
from collections import deque
from typing import Optional, List

# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def zigzagLevelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        result = []
        if not root:
            return result

        queue = deque([root])
        left_to_right = True  # Flag to determine traversal direction for the current level

        while queue:
            level_size = len(queue)
            current_level_nodes = []
            
            # Process all nodes at the current level
            for _ in range(level_size):
                node = queue.popleft()
                current_level_nodes.append(node.val)
                
                # Add children to the queue for the next level
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            
            # Apply zigzag order: Reverse if it's a right-to-left level
            if not left_to_right:
                current_level_nodes.reverse()
            
            result.append(current_level_nodes)
            
            # Toggle the flag for the next level
            left_to_right = not left_to_right
            
        return result
```

```python
# Construct this tree:
#       1
#      / \
#     2   3
#      \    \
#       5    4

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.right = TreeNode(5)
root.right.right = TreeNode(4)

sol = Solution()
result = sol.zigzagLevelOrder(root)
print("Right side view:", result)
```
