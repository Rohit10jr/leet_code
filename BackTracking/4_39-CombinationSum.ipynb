**39. Combination Sum**

Given an array of distinct integers candidates and a target integer target, return a list of all unique combinations of candidates where the chosen numbers sum to target. You may return the combinations in any order.

The same number may be chosen from candidates an unlimited number of times. Two combinations are unique if the frequency of at least one of the chosen numbers is different.

The test cases are generated such that the number of unique combinations that sum up to target is less than 150 combinations for the given input.


Example 1:

    Input: candidates = [2,3,6,7], target = 7
    Output: [[2,2,3],[7]]
    Explanation:
    2 and 3 are candidates, and 2 + 2 + 3 = 7. Note that 2 can be used multiple times.
    7 is a candidate, and 7 = 7.
    These are the only two combinations.

Example 2:

    Input: candidates = [2,3,5], target = 8
    Output: [[2,2,2,2],[2,3,3],[3,5]]

```python
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        result = []

        def backtrack(start, current, total):
            # If total equals the target, we found a valid combination
            if total == target:
                result.append(current[:])
                return
            
            # If total exceeds target, stop exploring
            if total > target:
                return

            # Try each candidate starting from 'start'
            for i in range(start, len(candidates)):
                current.append(candidates[i])  # choose

                backtrack(i, current, total + candidates[i])  # explore further

                current.pop()  # undo the choice (backtrack)

        backtrack(0, [], 0)
        return result
```

```python
class Solution1:
    def combinationSum(self, candidates, target):
        result = []

        def backtrack(start, current, total, depth_path):
            depth_str = ".".join(map(str, depth_path)) if depth_path else "ROOT"
            print(f"\n[{depth_str}] CALL → start={start}, current={current}, total={total}")

            # Base case: exact match
            if total == target:
                print(f"[{depth_str}] ✔ VALID → {current}")
                result.append(current[:])
                return

            # Base case: too large
            if total > target:
                print(f"[{depth_str}] ✘ total {total} exceeded target → BACKTRACK")
                return

            # Loop through candidates
            for i in range(start, len(candidates)):
                num = candidates[i]
                print(f"[{depth_str}] Trying {num} at index {i}")

                # Choose
                current.append(num)
                print(f"[{depth_str}]   → choose {num}, current={current}")

                # Go deeper with updated depth: add (i+1) to path
                backtrack(i, current, total + num, depth_path + [i + 1])

                # Backtrack
                removed = current.pop()
                print(f"[{depth_str}]   ← un-choose {removed}, current={current}")

        backtrack(0, [], 0, [])
        return result
```
