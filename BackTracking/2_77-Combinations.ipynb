**77. Combinations**

Given two integers n and k, return all possible combinations of k numbers chosen from the range [1, n].

You may return the answer in any order.

 
Example 1:

    Input: n = 4, k = 2
    Output: [[1,2],[1,3],[1,4],[2,3],[2,4],[3,4]]
    
- Explanation: There are 4 choose 2 = 6 total combinations. Note that combinations are unordered, i.e., [1,2] and [2,1] are considered to be the same combination.

Example 2:

    Input: n = 1, k = 1
    Output: [[1]]
    
- Explanation: There is 1 choose 1 = 1 total combination.

```python
from typing import List

class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        res = []                    # final list of combinations
        path = []                   # current partial combination

        def backtrack(start: int):
            # If the current path has k numbers, add a copy to results
            if len(path) == k:
                res.append(path.copy())
                return

            # numbers still needed
            need = k - len(path)
            # i can go from 'start' up to 'n - need + 1' (prune impossible branches)
            # reason: if i > n - need + 1, there aren't enough numbers left to fill 'need' slots
            for i in range(start, n - need + 2):  # +2 because range end is exclusive
                path.append(i)         # choose i
                backtrack(i + 1)      # choose remaining numbers only from i+1..n
                path.pop()            # undo (backtrack)

        backtrack(1)                 # start choosing from 1
        return res
```

```python
sol = Solution()

# Example 1
n = 4
k = 2
print(sol.combine(n, k))   # Expected: [[1,2],[1,3],[1,4],[2,3],[2,4],[3,4]]

# Example 2
n = 1
k = 1
print(sol.combine(n, k))   # Expected: [[1]]
```
