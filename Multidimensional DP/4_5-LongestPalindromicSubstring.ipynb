**5. Longest Palindromic Substring**

Given a string s, return the longest palindromic substring in s.


Example 1:

    Input: s = "babad"
    Output: "bab"
    Explanation: "aba" is also a valid answer.

Example 2:

    Input: s = "cbbd"
    Output: "bb"

```python
class Solution(object):
    def longestPalindrome(self, s):
        """
        :type s: str
        :rtype: str
        """
        def expand_from_center(left, right):
            # Expand outward while characters match and bounds are valid
            while left >= 0 and right < len(s) and s[left] == s[right]:
                left -= 1
                right += 1
            # Return the palindrome substring after expansion
            return s[left + 1:right]

        longest = ""
        for i in range(len(s)):
            # Odd length palindrome (centered at i)
            odd = expand_from_center(i, i)
            # Even length palindrome (centered between i and i+1)
            even = expand_from_center(i, i + 1)
            
            # Take the longer of the two
            current_longest = odd if len(odd) > len(even) else even
            
            # Update result if a longer palindrome is found
            if len(current_longest) > len(longest):
                longest = current_longest
        
        return longest
```

```python
sol = Solution()
# s = "babad"
s = "cbbd"
sol.longestPalindrome(s)
```
