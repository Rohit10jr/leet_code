**15. 3Sum**

Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.
Notice that the solution set must not contain duplicate triplets.


Example 1:

Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]

Explanation: 

nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
The distinct triplets are [-1,0,1] and [-1,-1,2].

Notice that the order of the output and the order of the triplets does not matter.

Example 2:

Input: nums = [0,1,1]
Output: []

Explanation: The only possible triplet does not sum up to 0.

```python
class Solution:
    def threeSum(self, nums):
        nums.sort()  # Step 1: Sort the array
        result = []
        n = len(nums)

        for i in range(n):
            # Step 2: Skip duplicate numbers for the first number
            if i > 0 and nums[i] == nums[i - 1]:
                continue

            # Step 3: Use two pointers
            left = i + 1
            right = n - 1

            while left < right:
                total = nums[i] + nums[left] + nums[right]

                if total == 0:
                    result.append([nums[i], nums[left], nums[right]])

                    # Skip duplicates for second and third numbers
                    while left < right and nums[left] == nums[left + 1]:
                        left += 1
                    while left < right and nums[right] == nums[right - 1]:
                        right -= 1

                    left += 1
                    right -= 1

                elif total < 0:
                    # Sum is too small; we need a larger number.
                    # Since array is sorted, move left to a bigger number.
                    left += 1
                else:
                    # Sum is too big; we need a smaller number.
                    # Since array is sorted, move right to a smaller number.
                    right -= 1

        return result
```

```python
# nums = [-1, 0, 1, 2,]a
nums = [-4, -1, -1, 0, 1, 2]
sol = Solution()
print(sol.threeSum(nums))
```
