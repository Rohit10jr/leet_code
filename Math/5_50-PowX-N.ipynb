**50. Pow(x, n)**

Implement pow(x, n), which calculates x raised to the power n (i.e., xn).

Example 1:

    Input: x = 2.00000, n = 10
    Output: 1024.00000

Example 2:

    Input: x = 2.10000, n = 3
    Output: 9.26100

Example 3:

    Input: x = 2.00000, n = -2
    Output: 0.25000

Explanation: 2-2 = 1/22 = 1/4 = 0.25

```python
# Binary exponentiation (exponentiation by squaring) approach
class Solution:
    def myPow(self, x: float, n: int) -> float:
        # Handle the case where n is negative
        if n < 0:
            x = 1 / x
            n = -n

        res = 1.0 # Initialize the result
        print("Initial values")
        print("x =", x, "n =", n, "res =", res)
        while n > 0:
            print('-'*30)
            # If n is odd, multiply x with the current result
            if n % 2 == 1: 
                res *= x
            
            x *= x # Square x for the next iteration
            n //= 2 # Halve n
            
            print("x =", x, "n =", n, "res =", res)

        return res
```

```python
class Solution2(object):
    def myPow(self, x, n):
        """
        :type x: float
        :type n: int
        :rtype: float
        """
        result = 1
        abs_n = abs(n)
        while abs_n:
            if abs_n & 1:
                result *= x
            abs_n >>= 1
            x *= x

        return 1 / result if n < 0 else result
```

```python

# Example Usage:
solution = Solution()
print(f"pow(2.0, 10) = {solution.myPow(2.0, 3)}")        # Output: 1024.0
print(f"pow(2 b.1, 3) = {solution.myPow(2.1, 3)}")          # Output: 9.261
print(f"pow(2.0, -2) = {solution.myPow(2.0, -2)}")        # Output: 0.25
print(f"pow(0.0, 1) = {solution.myPow(0.0, 1)}")          # Output: 0.0
print(f"pow(1.0, 0) = {solution.myPow(1.0, 0)}")          # Output: 1.0
print(f"pow(0.00001, 2147483647) = {solution.myPow(0.00001, 2147483647)}") # Should be very close to 0
print(f"pow(1.00000, -2147483648) = {solution.myPow(1.00000, -2147483648)}") # Should be 1.0
```
