**172. Factorial Trailing Zeroes**

Given an integer n, return the number of trailing zeroes in n!.

Note that n! = n * (n - 1) * (n - 2) * ... * 3 * 2 * 1. 

Example 1:

    Input: n = 3
    Output: 0

Explanation: 3! = 6, no trailing zero.

Example 2:

    Input: n = 5
    Output: 1

Explanation: 5! = 120, one trailing zero.

```python
class Solution:
    def trailingZeroes(self, n: int) -> int:
        count = 0
        # Keep dividing n by 5 and adding the quotient to count
        # This implicitly handles multiples of 25, 125, etc.
        # For example, if n = 26:
        # First iteration: count += 26 // 5 = 5. n becomes 5.
        # Second iteration: count += 5 // 5 = 1. n becomes 1.
        # Third iteration: count += 1 // 5 = 0. n becomes 0. Loop ends.
        # Total count = 5 + 1 = 6.
        while n > 0:
            n //= 5  # Integer division
            count += n
        return count
```

```python
solution = Solution()
# Example Usage:
print(f"Trailing zeroes in 0! = {solution.trailingZeroes(0)}")   # Output: 0 (0! = 1, no zeroes)
solution = Solution()
print(f"Trailing zeroes in 3! = {solution.trailingZeroes(3)}")   # Output: 0
print(f"Trailing zeroes in 5! = {solution.trailingZeroes(5)}")   # Output: 1
print(f"Trailing zeroes in 10! = {solution.trailingZeroes(10)}") # Output: 2 (10! = 3,628,800)
print(f"Trailing zeroes in 25! = {solution.trailingZeroes(25)}") # Output: 6
print(f"Trailing zeroes in 26! = {solution.trailingZeroes(26)}") # Output: 6
```
