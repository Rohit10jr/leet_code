**49. Group Anagrams**

Given an array of strings strs, group the anagrams together. You can return the answer in any order.

 

Example 1:

    Input: strs = ["eat","tea","tan","ate","nat","bat"]
    Output: [["bat"],["nat","tan"],["ate","eat","tea"]]

Example 2:

    Input: strs = [""]
    Output: [[""]]

Example 3:

    Input: strs = ["a"]
    Output: [["a"]]

```python
from collections import defaultdict

class Solution:
    def groupAnagrams(self, strs: list[str]) -> list[list[str]]:
        anagram_groups = defaultdict(list)

        for word in strs:
            # Create a count array for 26 lowercase English letters
            counts = [0] * 26
            for char in word:
                counts[ord(char) - ord('a')] += 1
            
            # Convert the counts array to a tuple to use as a hashable key
            # Example: "eat" -> (1, 0, 0, 0, 1, 0, ... , 1, ...)
            canonical_key_tuple = tuple(counts) 
            
            # Add the original word to the list associated with this key
            anagram_groups[canonical_key_tuple].append(word)
        
        return list(anagram_groups.values())

        
class Solution2(object):
    def groupAnagrams(self, strs):
        """
        :type strs: List[str]
        :rtype: List[List[str]]
        """
        anagram_map = defaultdict(list)

        for word in strs:
            # Sort the word to get the key
            key = ''.join(sorted(word)) # This correctly creates the canonical key
            anagram_map[key].append(word)

        return list(anagram_map.values())
```

```python
# Example 1 
sol = Solution()
strs = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(sol.groupAnagrams(strs))

# Example 2
sol2 = Solution2()
strs = ["eat", "tea", "tan", "ate", "nat", "bat", "cat", "tab"]
print(sol2.groupAnagrams(strs))
```
