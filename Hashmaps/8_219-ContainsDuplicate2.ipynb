**219. Contains Duplicate II**

Given an integer array nums and an integer k, return true if there are two distinct indices i and j in the array such that nums[i] == nums[j] and abs(i - j) <= k.


Example 1:

    Input: nums = [1,2,3,1], k = 3
    Output: true

Example 2:

    Input: nums = [1,0,1,1], k = 1
    Output: true

Example 3:

    Input: nums = [1,2,3,1,2,3], k = 2
    Output: false

```python
# Time complexity is O(N), the space complexity is O(N)
class Solution:
    def containsNearbyDuplicate(self, nums: list[int], k: int) -> bool:
        # {number: latest_index}
        num_to_index = {}

        # Iterate through the array with both index (i) and value (num)
        for i, num in enumerate(nums):
            # Check if the current number is already in our map
            if num in num_to_index:
                # If it is, calculate the absolute difference between the current index
                # and the previously stored index for this number.
                prev_index = num_to_index[num]
                if abs(i - prev_index) <= k:
                    # If the difference is within k, we found our pair!
                    return True
            
            # Update the map with the current number's index.
            # This ensures we always store the *latest* seen index for each number.
            num_to_index[num] = i
        
        return False
```

```python
sol = Solution()

# Exampls 1
nums = [1, 0, 1, 1] 
k = 1
print(sol.containsNearbyDuplicate(nums, k))

# Exampls 2
nums = [1, 0, 2, 1] 
k = 3
print(sol.containsNearbyDuplicate(nums, k))
```
