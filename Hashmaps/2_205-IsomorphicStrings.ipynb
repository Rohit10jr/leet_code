
**205. Isomorphic Strings**

Given two strings s and t, determine if they are isomorphic.

Two strings s and t are isomorphic if the characters in s can be replaced to get t.

All occurrences of a character must be replaced with another character while preserving the order of characters. No two characters may map to the same character, but a character may map to itself.

 
Example 1:

    Input: s = "egg", t = "add"
    Output: true

Example 2:

    Input: s = "foo", t = "bar"
    Output: false

```python
class Solution:
    def isIsomorphic(self, s: str, t: str) -> bool:
        # Problem statement guarantees t.length == s.length, but good practice to check
        if len(s) != len(t):
            return False

        # Map characters from s to t
        s_to_t_map = {}
        # Map characters from t to s (to ensure bi-directional unique mapping)
        t_to_s_map = {}

        for i in range(len(s)):
            char_s = s[i]
            char_t = t[i]

            # Check mapping from s to t
            if char_s in s_to_t_map:
                # If char_s was already mapped, ensure it maps to the same char_t
                if s_to_t_map[char_s] != char_t:
                    return False
            else:
                # If char_s is new, add its mapping
                s_to_t_map[char_s] = char_t
            
            # Check mapping from t to s
            if char_t in t_to_s_map:
                # If char_t was already mapped, ensure it maps back to the same char_s
                if t_to_s_map[char_t] != char_s:
                    return False
            else:
                # If char_t is new, add its mapping
                t_to_s_map[char_t] = char_s
        
        # If the loop completes, all mappings are consistent and unique in both directions
        return True
        
class Solution2:
    def isIsomorphic(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False  # Different lengths can't be isomorphic

        mapping_s_to_t = {}
        mapped_chars_in_t = set()

        for char_s, char_t in zip(s, t):
            if char_s in mapping_s_to_t:
                # If we’ve seen char_s, it must always map to the same char_t
                if mapping_s_to_t[char_s] != char_t:
                    return False
            else:
                # If char_t is already mapped by another char_s → not one-to-one
                if char_t in mapped_chars_in_t:
                    return False

                # Create the mapping
                mapping_s_to_t[char_s] = char_t
                mapped_chars_in_t.add(char_t)

        return True
```

```python
# Example 
sol = Solution()
s = "egg" 
t = "add"
print(sol.isIsomorphic(s, t)) # Output: true

# Example 2:
sol2 = Solution2()
s = "foo"
t = "bar"
print(sol2.isIsomorphic(s, t)) # Output: false
```
