**383. Ransom Note**

Given two strings ransomNote and magazine, return true if ransomNote can be constructed by using the letters from magazine and false otherwise.

Each letter in magazine can only be used once in ransomNote.

Example 1:

    Input: ransomNote = "a", magazine = "b", 
    Output: false

Example 2:

    Input: ransomNote = "aa", magazine = "ab", 
    Output: false
 

```python
# Time Complexity: O(m+n), Space Complexity: O(1) 
from collections import Counter

class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        ransom_count = Counter(ransomNote)
        magazine_count = Counter(magazine)

        # For each character and its required count in ransomNote
        for char, count in ransom_count.items():
            # If magazine doesn't have enough of that character, return False
            if magazine_count[char] < count:
                return False
        return True


class Solution2:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        magazine_counts = Counter(magazine)

        # Go through each character in ransomNote
        for char in ransomNote:
            # If magazine has that character available
            if magazine_counts[char] > 0:
                # Use the character (decrease its count)
                magazine_counts[char] -= 1
            else:
                # Character not available or already used up â†’ can't construct
                return False

        return True
```

```python
sol = Solution()
sol2 = Solution2()

# Example 1
ransomNote = "aa"
magazine = "aabc"

# Example 2
ransomNote2 = "bav"
magazine2 = "aabc"

print(sol.canConstruct(ransomNote, magazine))
print(sol2.canConstruct(ransomNote2, magazine2))
```

```python
# Counter example 
from collections import Counter

fruits = ["apple", "banana", "orange", "apple", "grape", "banana", "apple", "kiwi"]

fruit_counts = Counter(fruits)

print("--- 1. Using Counter() ---")
print(f"Original list of fruits: {fruits}")
print(f"Fruit counts: {fruit_counts}")
# Output: Fruit counts: Counter({'apple': 3, 'banana': 2, 'orange': 1, 'grape': 1, 'kiwi': 1})

print("\n--- 2. Accessing individual counts ---")
print(f"Number of 'apple': {fruit_counts['apple']}")  
print(f"Number of 'banana': {fruit_counts['banana']}")
print(f"Number of 'mango' (not in list): {fruit_counts['mango']}")# Counter returns 0 for missing keys, not KeyError!

print("\n--- 3. Using items() to iterate through counts ---")
print("Fruit and their counts:")
for fruit, count in fruit_counts.items():
    print(f"- {fruit}: {count}")
# Output:
# - apple: 3
# - banana: 2
# - orange: 1
# - grape: 1
# - kiwi: 1

print("\n--- 4. Using keys() and values() ---")
print(f"Just the fruits (keys): {fruit_counts.keys()}")
print(f"Just the counts (values): {fruit_counts.values()}")
# Output (order might vary for keys/values):
# Just the fruits (keys): dict_keys(['apple', 'banana', 'orange', 'grape', 'kiwi'])
# Just the counts (values): dict_values([3, 2, 1, 1, 1])

print("\n--- 5. Most common elements ---")
print(f"Top 2 most common fruits: {fruit_counts.most_common(2)}")
# Output: Top 2 most common fruits: [('apple', 3), ('banana', 2)]

print("\n--- 6. Updating a Counter ---")
more_fruits = ["pear", "apple", "pear"]
fruit_counts.update(more_fruits)
print(f"After adding more fruits: {fruit_counts}")
# Output: After adding more fruits: Counter({'apple': 4, 'banana': 2, 'pear': 2, 'orange': 1, 'grape': 1, 'kiwi': 1})

print("\n--- 7. Subtracting from a Counter ---")
eaten_fruits = ["apple", "banana"]
fruit_counts.subtract(eaten_fruits)
print(f"After eating some fruits: {fruit_counts}")
# Output: After eating some fruits: Counter({'apple': 3, 'banana': 1, 'pear': 2, 'orange': 1, 'grape': 1, 'kiwi': 1})
```
