242. Valid Anagram

Given two strings s and t, return true if t is an anagram of s, and false otherwise.

 
Example 1:

    Input: s = "anagram", t = "nagaram"
    Output: true

Example 2:

    Input: s = "rat", t = "car"
    Output: false

```python
from collections import Counter

class Solution2:
    def isAnagram(self, s: str, t: str) -> bool:
        return Counter(s) == Counter(t)

class Solution3(object):
    def isAnagram(self, s, t):
        return sorted(s) == sorted(t)
```

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        # 1. Check if lengths are different. If so, they cannot be anagrams.
        if len(s) != len(t):
            return False

        # 2. Initialize a frequency array for 26 lowercase English letters.
        # Each index corresponds to a letter: 0 for 'a', 1 for 'b', etc.
        counts = [0] * 26

        # 3. Iterate through string 's' and increment counts
        for char_s in s:
            # ord(char) gives the ASCII value. Subtract ord('a') to get 0-25 index.
            counts[ord(char_s) - ord('a')] += 1

        # 4. Iterate through string 't' and decrement counts
        for char_t in t:
            counts[ord(char_t) - ord('a')] -= 1
            # Optional early exit: if a count drops below zero, 
            # it means 't' has more of a character than 's', so it can't be an anagram.
            # This can sometimes save a little time but doesn't change worst-case complexity.
            # if counts[ord(char_t) - ord('a')] < 0:
            #    return False

        # 5. After processing both strings, check if all counts are zero.
        # If any count is non-zero, it means character frequencies don't match.
        for count in counts:
            if count != 0:
                return False
        
        # If all counts are zero, they are anagrams.
        return True
```

```python
# Example1 
sol = Solution()
s = "anagram"
t = "nagaram"
print(sol.isAnagram(s, t))

# Example2 
sol2 = Solution2()
s = "anagram"
t = "nagaram"
print(sol2.isAnagram(s, t))

# Example3 
sol3 = Solution3()
s = "anagram"
t = "nagaram"
print(sol3.isAnagram(s, t))
```
