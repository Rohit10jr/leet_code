**4. Median of Two Sorted Arrays**

Given two sorted arrays nums1 and nums2 of size m and n respectively, return the median of the two sorted arrays.

The overall run time complexity should be O(log (m+n)).

 

Example 1:

    Input: nums1 = [1,3], nums2 = [2]
    Output: 2.00000

Explanation: merged array = [1,2,3] and median is 2.

Example 2:

    Input: nums1 = [1,2], nums2 = [3,4]
    Output: 2.50000
    
Explanation: merged array = [1,2,3,4] and median is (2 + 3) / 2 = 2.5.

```python
class Solution:
    def findMedianSortedArrays(self, nums1, nums2):
        # Merge the arrays into a single sorted array.
        merged = nums1 + nums2

        # Sort the merged array.
        merged.sort()

        # Calculate the total number of elements in the merged array.
        total = len(merged)

        if total % 2 == 1:
            # If the total number of elements is odd, return the middle element as the median.
            return float(merged[total // 2])
        else:
            # If the total number of elements is even, calculate the average of the two middle elements as the median.
            middle1 = merged[total // 2 - 1]
            middle2 = merged[total // 2]
            return (float(middle1) + float(middle2)) / 2.0
```

```python

class Solution:
    def findMedianSortedArrays(self, nums1, nums2):
        m = len(nums1)
        n = len(nums2)

        # Ensure nums1 is the shorter array for binary search optimization.
        if m > n:
            nums1, nums2 = nums2, nums1
            m, n = n, m

        low = 0
        high = m

        total_left_half = (m + n + 1) // 2

        while low <= high:
            i = low + (high - low) // 2
            j = total_left_half - i

            # Use float('-inf') and float('inf') for boundary conditions
            nums1_left_max = float('-inf') if i == 0 else nums1[i - 1]
            nums1_right_min = float('inf') if i == m else nums1[i]

            nums2_left_max = float('-inf') if j == 0 else nums2[j - 1]
            nums2_right_min = float('inf') if j == n else nums2[j]

            if nums1_left_max <= nums2_right_min and nums2_left_max <= nums1_right_min:
                if (m + n) % 2 == 1:
                    return float(max(nums1_left_max, nums2_left_max))
                else:
                    left_max = max(nums1_left_max, nums2_left_max)
                    right_min = min(nums1_right_min, nums2_right_min)
                    return (float(left_max + right_min)) / 2.0
            elif nums1_left_max > nums2_right_min:
                high = i - 1
            else: # nums2_left_max > nums1_right_min
                low = i + 1

        return 0.0
```

```python
class Solution:
    def findMedianSortedArrays(self, nums1, nums2, method='binary') -> float:
        # Switch between methods here:
        if method == 'merge':
            return self.findMedian_merge_sort(nums1, nums2)
        elif method == 'compact':
            return self.findMedian_one_liner(nums1, nums2)
        elif method == 'binary':
            return self.findMedian_binary_search(nums1, nums2)
        else:
            raise ValueError("Invalid method. Choose from: 'merge', 'compact', 'binary'.")

    def findMedian_merge_sort(self, nums1, nums2):
        merged = nums1 + nums2
        merged.sort()
        total = len(merged)

        if total % 2 == 1:
            return float(merged[total // 2])
        else:
            mid1 = merged[total // 2 - 1]
            mid2 = merged[total // 2]
            return (mid1 + mid2) / 2.0

    def findMedian_one_liner(self, nums1, nums2):
        nums = sorted(nums1 + nums2)
        l = len(nums)
        mid = l // 2
        return float(nums[mid] if l % 2 else (nums[mid - 1] + nums[mid]) / 2.0)

    def findMedian_binary_search(self, nums1, nums2):
        m, n = len(nums1), len(nums2)
        if m > n:
            nums1, nums2, m, n = nums2, nums1, n, m

        low, high = 0, m
        half_len = (m + n + 1) // 2

        while low <= high:
            i = (low + high) // 2
            j = half_len - i

            max_left_1 = float('-inf') if i == 0 else nums1[i - 1]
            min_right_1 = float('inf') if i == m else nums1[i]

            max_left_2 = float('-inf') if j == 0 else nums2[j - 1]
            min_right_2 = float('inf') if j == n else nums2[j]

            if max_left_1 <= min_right_2 and max_left_2 <= min_right_1:
                if (m + n) % 2 == 1:
                    return float(max(max_left_1, max_left_2))
                else:
                    return (max(max_left_1, max_left_2) + min(min_right_1, min_right_2)) / 2.0
            elif max_left_1 > min_right_2:
                high = i - 1
            else:
                low = i + 1

        return 0.0  # Fallback (should never be hit if input is valid)
```

```python
s = Solution()

# Use binary search method (fastest, O(log(min(m,n))))
print(s.findMedianSortedArrays([1, 3], [2], method='binary'))   # Output: 2.0

# Use merge + sort method
print(s.findMedianSortedArrays([1, 2], [3, 4], method='merge'))  # Output: 2.5

# Use compact one-liner sort version
print(s.findMedianSortedArrays([1, 2], [3, 4], method='compact'))  # Output: 2.5
```
