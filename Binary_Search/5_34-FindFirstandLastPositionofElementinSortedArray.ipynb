**34. Find First and Last Position of Element in Sorted Array**

Given an array of integers nums sorted in non-decreasing order, find the starting and ending position of a given target value.

If target is not found in the array, return [-1, -1].

You must write an algorithm with O(log n) runtime complexity.


Example 1:

    Input: nums = [5,7,7,8,8,10], target = 8
    Output: [3,4]

Example 2:

    Input: nums = [5,7,7,8,8,10], target = 6
    Output: [-1,-1]

```python
class Solution:
    def searchRange(self, nums: list[int], target: int) -> list[int]:
        def find_boundary(nums: list[int], target: int, find_first: bool) -> int:
            """
            Performs a modified binary search to find either the first or last occurrence of target.

            Args:
                nums: The sorted array.
                target: The value to search for.
                find_first: If True, search for the first occurrence. If False, search for the last.

            Returns:
                The index of the first/last occurrence, or -1 if not found.
            """
            # print("="*5, "finding first position" if find_first else "finding second position", "="*5)
            low = 0
            high = len(nums) - 1
            ans = -1
            i = 0
            print("="*5,"Outiside while loop", "="*5)
            print(f"nums={nums}, target={target}, find_first={find_first}, low={low}, high={high}, ans={ans}")

            while low <= high:
                mid = low + (high - low) // 2

                if nums[mid] == target:
                    ans = mid  # Found a potential answer
                    if find_first:
                        high = mid - 1  # Try to find an earlier occurrence
                    else:
                        low = mid + 1   # Try to find a later occurrence
                elif nums[mid] < target:
                    low = mid + 1
                else:  # nums[mid] > target
                    high = mid - 1
                i += 1 # iteration count
                print(f"--- inside while loop, iteration {i} ---")
                print(f"mid={mid}, low={low}, high={high}, ans={ans}" if mid else "mid is None")
            return ans

        print("<<"*5,"find first position", ">>"*5)
        first_pos = find_boundary(nums, target, True)
        print("<<"*5,"find last position", ">>"*5)
        last_pos = find_boundary(nums, target, False)

        return [first_pos, last_pos]
```

```python
s = Solution()
print(s.searchRange([5,7,7,8,8,10], 8))   # Output: [3, 4]
# print(s.searchRange([5,7,7,8,8,10], 6))   # Output: [-1, -1]
# print(s.searchRange([], 0))              # Output: [-1, -1]
```
