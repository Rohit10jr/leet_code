**35. Search Insert Position**

Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

You must write an algorithm with O(log n) runtime complexity.

 

Example 1:

    Input: nums = [1,3,5,6], target = 5
    Output: 2

Example 2:

    Input: nums = [1,3,5,6], target = 2
    Output: 1

**Whatâ€™s Binary Search?**

Binary search works only on sorted arrays.

Basic idea:

1. Look at the middle of the array

2. If it's what you want, you're done

3. If it's smaller than your target, discard the left half

4. If it's larger, discard the right half

5. Repeat on the remaining half

```python
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        left, right = 0, len(nums) - 1
        
        while left <= right:
            mid = (left + right) // 2
            
            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                left = mid + 1
            else: # nums[mid] > target
                right = mid - 1

        # If the loop finishes and target was not found.
        # 'low' now represents the insertion point.
        return left
```

```python
sol = Solution()

# Example Usage:
print(f"Example 1: Input: nums = [1,3,5,6], target = 5, Output: {sol.searchInsert([1,3,5,6], 5)}")
print(f"Example 2: Input: nums = [1,3,5,6], target = 2, Output: {sol.searchInsert([1,3,5,6], 2)}")
print(f"Example 3: Input: nums = [1,3,5,6], target = 7, Output: {sol.searchInsert([1,3,5,6], 7)}") # Target greater than all
print(f"Example 4: Input: nums = [1,3,5,6], target = 0, Output: {sol.searchInsert([1,3,5,6], 0)}") # Target smaller than all
print(f"Example 5: Input: nums = [], target = 5, Output: {sol.searchInsert([], 5)}")
```
