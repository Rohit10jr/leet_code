**190. Reverse bits**

Reverse bits of a given 32 bits unsigned integer.

*Note:*

Note that in some languages, such as Java, there is no unsigned integer type. In this case, both input and output will be given as a signed integer type. They should not affect your implementation, as the integer's internal binary representation is the same, whether it is signed or unsigned.
In Java, the compiler represents the signed integers using 2's complement notation. Therefore, in Example 2 above, the input represents the signed integer -3 and the output represents the signed integer -1073741825.
 

Example 1:

    Input: n = 00000010100101000001111010011100
    Output:    964176192 (00111001011110000010100101000000)

Explanation: The input binary string 00000010100101000001111010011100 represents the unsigned integer 43261596, so return 964176192 which its binary representation is 00111001011110000010100101000000.

```python
class Solution:
    def reverseBits(self, n: int) -> int:
        reversed_n = 0  # Initialize the result to 0

        # Iterate 32 times, processing each bit
        for i in range(32):
            # 1. Extract the Least Significant Bit (LSB) of n
            #    n & 1 will be 1 if the LSB is 1, otherwise 0
            bit = n & 1

            # 2. Shift reversed_n left by 1 to make space for the new bit
            #    This effectively moves all existing bits in reversed_n to the left
            reversed_n = reversed_n << 1

            # 3. Add the extracted bit to the LSB position of reversed_n
            #    Using bitwise OR (|) to set the bit
            reversed_n = reversed_n | bit

            # 4. Shift n right by 1 to process the next bit in the next iteration
            #    This effectively discards the LSB we just processed from n
            n = n >> 1

        return reversed_n
```

```python
solver = Solution()

# Example 1: n = 00000010100101000001111010011100 (decimal 43261596)
n1 = 43261596 
# Expected output: 964176192 (00111001011110000010100101000000)
print(f"Input: {n1} ({bin(n1)[2:].zfill(32)})")
output1 = solver.reverseBits(n1)
print(f"Output: {output1} ({bin(output1)[2:].zfill(32)})")
print("-" * 30)

# Example with a smaller number for easier tracing
# Input: 2 (binary 0...0010)
# Expected: 1073741824 (binary 0100...000)
n_small = 2
print(f"Input: {n_small} ({bin(n_small)[2:].zfill(32)})")
output_small = solver.reverseBits(n_small)
print(f"Output: {output_small} ({bin(output_small)[2:].zfill(32)})")
print("-" * 30)
```

**Bitwise Operators**

| Operator | Name           | Example (in binary)      | Description                                        |
|----------|----------------|-------------------------|-----------------------------------------------------|
| `&`      | AND            | `1101 & 1011 = 1001`    | 1 if both bits are 1                                |
| `\|`      | OR             | `1101 \| 1011 = 1111`   | 1 if at least one bit is 1                          |
| `^`      | XOR            | `1101 ^ 1011 = 0110`    | 1 if bits are different                             |
| `~`      | NOT (invert)   | `~1101 = ...11110010`   | Inverts all the bits (bitwise negation)             |
| `<<`     | Left Shift     | `0001 << 2 = 0100`      | Shifts bits left, adds zeros on the right           |
| `>>`     | Right Shift    | `1000 >> 2 = 0010`      | Shifts bits right, removes rightmost bits           |

```python
a = 13       # binary: 1101
b = 11       # binary: 1011

print(bin(a & b))   # 0b1001 → 9
print(bin(a | b))   # 0b1111 → 15
```
