191. Number of 1 

Given a positive integer n, write a function that returns the number of set bits in its binary representation (also known as the Hamming weight).
 

Example 1:

    Input: n = 11
    Output: 3

Explanation:
The input binary string 1011 has a total of three set bits.

Example 2:

    Input: n = 128
    Output: 1

Explanation:
The input binary string 10000000 has a total of one set bit.

```python
class Solution:
    def hammingWeight(self, n: int) -> int:
        """
        Calculates the number of set bits (Hamming weight) in a positive integer.
        Uses Brian Kernighan's algorithm (n & (n - 1)).

        Args:
            n: A positive integer.

        Returns:
            The number of set bits in its binary representation.
        """
        count = 0
        while n > 0:
            count += 1
            # This operation effectively turns off the least significant set bit
            n = n & (n - 1)
        return count
    
    def bit_manipulation(self, n: int) -> int:
        """
        Calculates the number of set bits (Hamming weight) in a positive integer
        by checking each bit iteratively from LSB to MSB.
        """
        count = 0
        # Loop 32 times for a 32-bit integer,
        # or while n > 0 for a generic positive integer
        for _ in range(32): # Or while n > 0:
            # Check if the least significant bit is 1
            if (n & 1) == 1:
                count += 1
            # Right-shift n by 1 to move to the next bit
            n = n >> 1
        return count

    
    def string_conversion(self, n: int) -> int:
        # 1. Convert integer to binary string, [2:] slices to remove the '0b' prefix.
        binary_str = bin(n)[2:]
        # 2. Count the occurrences of '1' in the binary string.
        count = binary_str.count('1')
        return count


```

```python

solver = Solution()

# Example 1: n = 11 (binary 1011) -> 3 set bits
n1 = 11
print(f"Input: n = {n1} (binary: {bin(n1)})")
print(f"Output: {solver.hammingWeight(n1)}") # Expected: 3
print("-" * 30)

# Example 2: n = 128 (binary 10000000) -> 1 set bit
n2 = 128
print(f"Input: n = {n2} (binary: {bin(n2)})")
print(f"Output: {solver.hammingWeight(n2)}") # Expected: 1
print("-" * 30)

# Example 3: n = 2147483645 (binary 01111111111111111111111111111101) -> 30 set bits
n3 = 2147483645
print(f"Input: n = {n3} (binary: {bin(n3)})")
print(f"Output: {solver.hammingWeight(n3)}") # Expected: 30
print("-" * 30)

# Example 4: n = 0 -> 0 set bits
n4 = 0
print(f"Input: n = {n4} (binary: {bin(n4)})")
print(f"Output: {solver.hammingWeight(n4)}") # Expected: 0
print("-" * 30)

# Example 5: n = 1 -> 1 set bit
n5 = 1
print(f"Input: n = {n5} (binary: {bin(n5)})")
print(f"Output: {solver.hammingWeight(n5)}") # Expected: 1
print("-" * 30)
```
