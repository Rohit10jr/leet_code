**201. Bitwise AND of Numbers Range**

Given two integers left and right that represent the range [left, right], return the bitwise AND of all numbers in this range, inclusive.

Example 1:

    Input: left = 5, right = 7
    Output: 4

Example 2:

    Input: left = 0, right = 0
    Output: 0

```python
class Solution:
    def rangeBitwiseAnd(self, left: int, right: int) -> int:
        shift_count = 0

        # Find the common prefix by right-shifting both numbers
        while left<right:
        # while left != right:
            left >>= 1
            right >>= 1
            shift_count += 1
        
        # Restore the common prefix to its original position
        return left << shift_count
```

```python

solver = Solution()

# Example 1: Input: left = 5, right = 7, Output: 4
# 5: 0101
# 6: 0110
# 7: 0111
# AND: 0100 = 4
print(f"Input: left = 5, right = 7")
print(f"Output: {solver.rangeBitwiseAnd(5, 7)}") # Expected: 4
print("-" * 30)

# Example 2: Input: left = 0, right = 0, Output: 0
print(f"Input: left = 0, right = 0")
print(f"Output: {solver.rangeBitwiseAnd(0, 0)}") # Expected: 0
print("-" * 30)

# Example 3: Input: left = 1, right = 2147483647 (max 32-bit signed int)
# Binary of 1 is ...001. Binary of max int is ...111.
# The only common prefix will be all zeros.
print(f"Input: left = 1, right = 2147483647")
print(f"Output: {solver.rangeBitwiseAnd(1, 2147483647)}") # Expected: 0
print("-" * 30)

# Example 4: Input: left = 8, right = 8
# 8: 1000
print(f"Input: left = 8, right = 8")
print(f"Output: {solver.rangeBitwiseAnd(8, 8)}") # Expected: 8
print("-" * 30)

# Example 5: Input: left = 9, right = 12
# 9:  1001
# 10: 1010
# 11: 1011
# 12: 1100
# Common prefix is 1000... (at the 3rd bit from MSB, they start differing)
#   The common prefix is '100', then the remaining bits will all become 0.
#   Result should be 8.
print(f"Input: left = 9, right = 12")
print(f"Output: {solver.rangeBitwiseAnd(9, 12)}") # Expected: 8
print("-" * 30)
```
