**137. Single Number II**

Given an integer array nums where every element appears three times except for one, which appears exactly once. Find the single element and return it.

You must implement a solution with a linear runtime complexity and use only constant extra space.
 

Example 1:

    Input: nums = [2,2,3,2]
    Output: 3

Example 2:

    Input: nums = [0,1,0,1,0,1,99]
    Output: 99

```python
class Solution:
    def singleNumber(self, nums: list[int]) -> int:
        single_number = 0

        # For a 32-bit integer
        for i in range(32):
            bit_sum = 0  
            # Iterate through each number in the array
            for num in nums:
                # Check if the i-th bit of the current number is set (1)
                # (num >> i) shifts the i-th bit to the LSB position & 1 isolates that LSB
                if (num >> i) & 1:
                    bit_sum += 1

            # If the sum of bits at this position is not a multiple of 3,
            # it means the single number must have a '1' at this bit position.
            if bit_sum % 3 != 0:
                # | performs a bitwise OR to set that bit
                single_number |= (1 << i)
        
        # If the 31st bit (most significant bit for a 32-bit signed int) is set,
        # it indicates a negative number in 2's complement.
        if single_number >= (1 << 31):
            single_number -= (1 << 32)
            
        return single_number
```

```python
class Solution(object):
    def singleNumber(self, nums):
        ones,twos = 0,0
        for num in nums:
            ones = (ones^num) & ~twos
            twos = (twos^num) & ~ones
        return ones
```

```python

solver = Solution()

# Example 1
nums2 = [0, 1, 0, 1, 0, 1, 99]
print(f"Input: {nums2}")
print(f"Output: {solver.singleNumber(nums2)}") # Expected: 99
print("-" * 30)

# Example 2: 
nums4 = [10**9, 7, 7, 7]
print(f"Input: {nums4}")
print(f"Output: {solver.singleNumber(nums4)}") # Expected: 10**9
print("-" * 30)

# Example 3: Contains negative numbers [-2,-2,3,-2] -> 3
nums_neg_ex = [-2, -2, 1, -2] 
print(f"Input: {nums_neg_ex}")
print(f"Output: {solver.singleNumber(nums_neg_ex)}") # Expected: 1
print("-" * 30)

nums_neg_ex2 = [-2, -2, 3, -2] # -2 (x3), 3 (x1)
print(f"Input: {nums_neg_ex2}")
print(f"Output: {solver.singleNumber(nums_neg_ex2)}") # Expected: 3
print("-" * 30)
```
